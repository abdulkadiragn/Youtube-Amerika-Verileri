```python
#video beğeni oranına göre gösterme
#iteritems degerleri ve indexleri alarak bir tupple'a yerleştirir
def begenOrani(likes,dislikes):
    begeniListesi=[]
    for key,value in list(likes.iteritems()):
        begeniListesi.append(value)
        
    dislikeListesi=[]
    for key,value in list(dislikes.iteritems()):
        dislikeListesi.append(value)
        
    likeDislike = list(zip(begeniListesi,dislikeListesi))
    oran = []
    for like,dislike in likeDislike:
        if(like + dislike ) == 0:
            oran.append(0)
        else:
            oran.append(like / (like + dislike))
    print(oran)
    return oran
data["begeni_orani"] = begenOrani(data["likes"],data["dislikes"])
```
